### Function to generate random numbers

```python
#defining a function to generate random numbers
def random_number_generator(lower_limit,upper_limit,size):
    """
    Function which generates random numbers between two given intervals,
    which takes lower limit,upper limit and size i.e total random numbers that need to generated 
    """
    randoms=[]                                  #initialzing a empty list
    count = lower_limit                         #initializing count to lower limit
    limit=upper_limit-lower_limit               
    step=1/(size/limit)                         #calculating the step size
    
    while count < upper_limit:
        randoms.append(count)                   #appending random numbers to list
        count += step                           #incrementing
    return randoms
```

### Testing the function

```python
#generating 20 random numbers between 1 and 5
random_number_generator(1,5,20)
```

### Plotting the random numbers


```python
#importing the necessary library
import matplotlib.pyplot as plt

#generating 100 random numbers between 10,30
random_numbers=random_number_generator(10,30,100)

#plotting the random numbers
plt.figure(figsize=(20,10))
plt.plot(random_numbers)
plt.show()
```

### Calculating the value of pi using Leibniz’s formula

```python

# Initializing the denominator
k = 1
 
# Initializing the value of pi
pi = 0

#iterating over the numbers(more the iterations accurate the value will be)
for i in random_number_generator(0,300,300):
 
    if i % 2 == 0:                 # elemnts with even index are positive
        pi+= 4 * 1/k
        
    else:
        pi-= 4 * 1/k                 # odd index elements are negative
 
    k += 2                           # denominator should be odd 
     
print(f'The value of pi is: {round(pi,2)}') #rounding off to two decimal places
```
